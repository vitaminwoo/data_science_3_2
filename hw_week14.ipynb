```python
import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras import initializers
import tensorflow as tf

#y = 1 + 3*x1 + 5*x2 ^ 2 + 10*x3 ^ 3 + 20*x4 ^ 4


def gen_sequential_model():
    tf.random.set_seed(42)
    model = Sequential([
            Input(4, name = 'input_layer'), 
            Dense(16, activation = 'relu', name = 'hidden_layer1', 
                  kernel_initializer=initializers.RandomNormal(mean=0.0, stddev=0.05, seed=42)),
            Dense(16, activation = 'relu', name = 'hidden_layer2', 
                  kernel_initializer=initializers.RandomNormal(mean=0.0, stddev=0.05, seed=42)),
            Dense(16, activation = 'relu', name = 'hidden_layer3', 
                  kernel_initializer=initializers.RandomNormal(mean=0.0, stddev=0.05, seed=42)),
            Dense(1, activation = 'relu', name = 'output_layer', 
                  kernel_initializer=initializers.RandomNormal(mean=0.0, stddev=0.05, seed=42))
            ])
    model.summary()
    #print(model.layers[0].get_weights())
    #print(model.layers[1].get_weights())
    #print(model.layers[2].get_weights())
    #print(model.layers[3].get_weights())
    
    model.compile(optimizer='sgd', loss='mse')
    
    return model

#y = a + b*x1 + c*x2 ^ 2 + d*x3 ^ 3 + e*x4 ^ 4

def gen_linear_regression_dateset(numofsamples=500, a=1, b=3, c=5, d=10, e=20):
    np.random.seed(42)
    X = np.random.rand(numofsamples, 4)
    X_power = X.copy()
    X_power = X_power.transpose()
    X_power[1] =X_power[1] ** 2
    X_power[2] =X_power[2] ** 3
    X_power[3] =X_power[3] ** 4
    X_power = X_power.transpose()
    #print(X)
    #print(X.shape)
    
    coef = np.array([b, c, d, e])
    bias = a
    #print(coef)
    #print(coef.shape)
    
    y = np.matmul(X_power, coef.transpose()) + bias
   # print(y)
   # print(y.shape)

    return X, y


def plot_loss_curve(history):

    import matplotlib.pyplot as plt

    plt.figure(figsize=(15, 10))

    plt.plot(history.history['loss'][1:])
    plt.plot(history.history['val_loss'][1:])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper right')
    plt.show()   

model = gen_sequential_model()

X, y = gen_linear_regression_dateset(numofsamples=12800) #16뉴런 최종본 (20배)

history = model.fit(X, y, epochs = 500, verbose = 2, validation_split=0.3)

plot_loss_curve(history)
print("train loss=", history.history['loss'][-1])  #loss 출력 (아랫줄은 validation에 대한 loss.)
print("test loss=", history.history['val_loss'][-1])
```

```python
def predict_new_sample(model, x, a=1, b=3, c=5, d=10, e=20):
    
    x = x.reshape(1,4) #1차원의 x를 1x4차원으로 수정.
    y_pred = model.predict(x)[0][0]
    
    y_actual = a + b*x[0][0] + c*(x[0][1]**2) + d*(x[0][2]**3) + e*(x[0][3]**4)
    
    print("y actual value = ", y_actual)
    print("y predicted value = ", y_pred)
    print("차이 = ", (y_actual-y_pred))
    print()
    
predict_new_sample(model, np.array([0.6, 0.3, 0.7, 0.3]))
predict_new_sample(model, np.array([0.1, 0.2, 0.3, 0.4]))
predict_new_sample(model, np.array([0.5, 0.6, 0.7, 0.8]))
predict_new_sample(model, np.array([0.4, 0.3, 0.2, 0.1]))
```
