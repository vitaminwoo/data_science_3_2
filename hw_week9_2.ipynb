```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pymysql
import time


def load_dbscore_data9():
    conn = pymysql.connect(host='localhost', user='root', password='---', db='datascience_week9', charset='utf8') 
    curs = conn.cursor(pymysql.cursors.DictCursor)
    
    sql = "select * from db_score"
    curs.execute(sql)
    
    data  = curs.fetchall()
    
    curs.close()
    conn.close()
    
    X = [ (t['attendance'], t['homework'], t['final'] ) for t in data ]
    #X = [ ( t['final'] ) for t in data ]
    X = np.array(X)
    
    y = [ (t['score']) for t in data]
    y = np.array(y)

    return X, y

X, y = load_dbscore_data9()
```

```python
import statsmodels.api as sm
X_const = sm.add_constant(X)

model = sm.OLS(y, X_const)
ls = model.fit()

print(ls.summary())


ls_c = ls.params[0]
ls_m1, ls_m2, ls_m3 = ls.params[1], ls.params[2], ls.params[3]

def gradient_descent_naive(X, y):

    
    epochs = 100000
    min_grad = 0.0001
    learning_rate = 0.001
    
    m_grad1, m_grad2, m_grad3 = 0.0, 0.0, 0.0
    c_grad1 = 0.0
    
    n = len(y)
    
    m_1, m_2, m_3 = 0.0, 0.0, 0.0
    c_1 = 0.0
    
    X1 = np.array([row[0] for row in X])
    X2 = np.array([row[1] for row in X])
    X3 = np.array([row[2] for row in X])

    for epoch in range(epochs):
        #y = m_1 * X1 + m_2 * X2 + m_3 * X3
        for i in range(n):
            y_pred = m_1 * X1[i] + m_2 * X2[i] + m_3 * X3[i] + c_1
            m_grad1 += 2*(y_pred - y[i]) * X1[i]
            m_grad2 += 2*(y_pred - y[i]) * X2[i]
            m_grad3 += 2*(y_pred - y[i]) * X3[i]
            c_grad1 += 2*(y_pred - y[i])

        m_grad1 /= n
        m_grad2 /= n
        m_grad3 /= n
        c_grad1 /= n
        
        m_1 = m_1 - learning_rate * m_grad1 
        m_2 = m_2 - learning_rate * m_grad2 
        m_3 = m_3 - learning_rate * m_grad3 
        c_1 = c_1 - learning_rate * c_grad1
        
       # m_grad1, m_grad2, m_grad3 = m_grad[0], m_grad[1], m_grad[2]
       # m_1, m_2, m_3 = m[0], m[1], m[2]
        
        #c_grad1 = c_grad[0]
        #c_1 = c[0]
        
        if ( epoch % 1000 == 0):
            print("epoch %d: m_grad1=%f, m_grad2=%f, m_grad3=%f, c_grad=%f, m1=%f, m2=%f, m3=%f, c=%f"
                  %(epoch, m_grad1, m_grad2, m_grad3, c_grad1, m_1, m_2, m_3, c_1) ) 

        if ( abs(m_grad1) < min_grad and abs(m_grad2) < min_grad and abs(m_grad3) < min_grad and abs(c_grad1) < min_grad ):
            break
        
        
    return m_1, m_2, m_3, c_1

start_time = time.time()
m_1, m_2, m_3, c_1 = gradient_descent_naive(X, y)
end_time = time.time()

print("%f seconds" %(end_time - start_time))


print("\n\nFinal:")
print("gdn_m1=%f, gdn_m2=%f, gdn_m3=%f,gdn_c=%f" %(m_1, m_2, m_3, c_1) )
print("ls_m1=%f, ls_m2=%f, ls_m3=%f,ls_c=%f" %(ls_m1, ls_m2, ls_m3, ls_c) )


```

```python
def gradient_descent_vectorized(X, y):
    epochs = 100000
    min_grad = 0.0001
    learning_rate = 0.001
    
    m_1, m_2, m_3 = 0.0, 0.0, 0.0
    c_1 = 0.0
    
    n = len(y)
    
    m_grad1, m_grad2, m_grad3 = 0.0, 0.0, 0.0
    c_grad1 = 0.0

        
    X1 = np.array([row[0] for row in X])
    X2 = np.array([row[1] for row in X])
    X3 = np.array([row[2] for row in X])

    for epoch in range(epochs):    
    
        y_pred = m_1 * X1 + m_2 * X2 + m_3 * X3 + c_1

        m_grad1 = (2 * (y_pred - y) * X1).sum()/n
        m_grad2 = (2 * (y_pred - y) * X2).sum()/n
        m_grad3 = (2 * (y_pred - y) * X3).sum()/n
        c_grad1 = (2 * (y_pred - y)).sum()/n
        
        m_1 = m_1 - learning_rate * m_grad1 
        m_2 = m_2 - learning_rate * m_grad2 
        m_3 = m_3 - learning_rate * m_grad3 
        c_1 = c_1 - learning_rate * c_grad1       

        
        if ( epoch % 1000 == 0):
            print("epoch %d: m_grad1=%f, m_grad2=%f, m_grad3=%f, c_grad=%f, m1=%f, m2=%f, m3=%f, c=%f"
                  %(epoch, m_grad1, m_grad2, m_grad3, c_grad1, m_1, m_2, m_3, c_1) ) 

        if ( abs(m_grad1) < min_grad and abs(m_grad2) < min_grad and abs(m_grad3) < min_grad and abs(c_grad1) < min_grad ):
            break
        
        
    return m_1, m_2, m_3, c_1

start_time = time.time()
m_1, m_2, m_3, c_1 = gradient_descent_vectorized(X, y)
end_time = time.time()

print("%f seconds" %(end_time - start_time))

print("\n\nFinal:")
print("gdn_m1=%f, gdn_m2=%f, gdn_m3=%f,gdn_c=%f" %(m_1, m_2, m_3, c_1) )
print("ls_m1=%f, ls_m2=%f, ls_m3=%f,ls_c=%f" %(ls_m1, ls_m2, ls_m3, ls_c) )
```
